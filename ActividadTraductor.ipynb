# Diccionario de Frases Hechas: Compilación y Análisis de Expresiones Idiomáticas

```python
diccionario = {
    "feliz": "happy",
    "triste": "sad",
    "rápido": "fast",
    "lento": "slow",
    "grande": "big",
    "pequeño": "small",
    "noche": "night",
    "día": "day",
    "amor": "love",
    "paz": "peace"
}
```

## Función para traducir palabras

```python
def translate(word):
  return diccionario.get(word, "no encontrada")
```

## Ejemplo de traducción

```python
word = "trabajo"
print(translate(word))
```
