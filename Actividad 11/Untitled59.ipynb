```python
import edge_tts
import asyncio

texto = """
Soy Erick Javier Sanchez Bernal
"""

async def generar():
    communicate = edge_tts.Communicate(
        texto,
        voice="es-MX-DaliaNeural"
    )

    await communicate.save("voz.mp3")

await generar()
```
